```python
import numpy as np
df = pd.read_stata(r"C:\Users\caval\Desktop\tp2\variables_filtradas.dta")

# Paso 1: Reemplazar valores vacíos por NaN
df['ch14'] = df['ch14'].replace({'': np.nan})

# Paso 2: Convertir todos los valores a strings para manejarlos de forma uniforme
df['ch14'] = df['ch14'].astype(str).str.strip()

# Paso 3: Función para limpiar cada valor
def convertir_a_entero(valor):
    if valor in ['98', '99', '98.0', '99.0']:
        return int(float(valor))  # mantener 98 y 99 como enteros
    try:
        # Convertimos valores como '01', '2.0', etc., a enteros
        return int(float(valor))
    except:
        return 0  # Si no se puede convertir, ponemos NaN

# Paso 4: Aplicar la función a toda la columna
df['ch14'] = df['ch14'].apply(convertir_a_entero)
```

```python
df['ch12'] = pd.to_numeric(df['ch12'], errors='coerce')
df['ch13'] = pd.to_numeric(df['ch13'], errors='coerce')
```

```python
import numpy as np
import pandas as pd

def calcular_anos_educacion(row):
    nivel = row['ch12']
    finalizo = row['ch13']
    ultimo_aprobado = row['ch14']

    if nivel == 9 or ultimo_aprobado in [98, 99]:
        return np.nan
    
    # Si finalizó el nivel
    if finalizo == 1:
        return {
            1: 0,
            2: 7,
            3: 9,
            4: 12,   # asumimos primaria + secundaria
            5: 12,
            6: 15,   # terciario
            7: 17,   # universitario
            8: 20,   # posgrado
        }.get(nivel, np.nan)

    # Si no lo finalizó
    if finalizo == 2:
        # Años previos al nivel actual
        previos = {
            1: 0,
            2: 0,
            3: 0,
            4: 9,
            5: 9,
            6: 12,
            7: 15,
            8: 17
        }.get(nivel, 0)

        # Validar que el año aprobado esté en rango
        if 1 <= ultimo_aprobado <= 9:
            return previos + ultimo_aprobado
        else:
            return np.nan

    return np.nan

# Asumiendo que tus datos están en un DataFrame llamado df:
df['anos_educacion'] = df.apply(calcular_anos_educacion, axis=1)

df.to_stata(r"C:\Users\caval\Desktop\tp2\creamos_variable.dta", write_index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar la base
df = pd.read_stata("hogares_final.dta")

# Función kernel uniforme (top-hat)
def kernel_uniform(x, data, h):
    n = len(data)
    dens = np.zeros_like(x)
    for xi in data:
        mask = np.abs(x - xi) <= h
        dens[mask] += 1 / (2 * h)
    return dens / n

# Años a comparar
años = [2004, 2024]

for año in años:
    # Filtrar datos del año
    df_año = df[df["ano4"] == año].copy()

    # Recorte de outliers
    p1 = df_año["ingreso_hogar"].quantile(0.01)
    p99 = df_año["ingreso_hogar"].quantile(0.99)
    df_año = df_año[(df_año["ingreso_hogar"] >= p1) & (df_año["ingreso_hogar"] <= p99)]

    # Separar grupos
    ocup = df_año[df_año["algun_desocupado"] == 0]["ingreso_hogar"]
    desocup = df_año[df_año["algun_desocupado"] == 1]["ingreso_hogar"]

    # Ancho de banda
    h = 0.1 * df_año["ingreso_hogar"].std()

    # Rango para evaluar
    x_vals = np.linspace(df_año["ingreso_hogar"].min(), df_año["ingreso_hogar"].max(), 500)

    # Estimar densidades
    dens_ocup = kernel_uniform(x_vals, ocup.values, h)
    dens_desocup = kernel_uniform(x_vals, desocup.values, h)

    # Gráfico
    plt.figure(figsize=(10, 5))
    plt.plot(x_vals, dens_ocup, label="Todos ocupados", color="#9467bd", linewidth=2.2)  # violeta
    plt.plot(x_vals, dens_desocup, label="Al menos un desocupado", color="#ff7f0e", linewidth=2.2)  # naranja
    plt.title(f"Densidad del ingreso por hogar - Año {año} (kernel uniforme)", fontsize=13)
    plt.xlabel("Ingreso del hogar", fontsize=11)
    plt.ylabel("Densidad estimada", fontsize=11)
    plt.legend()
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar base
df = pd.read_stata("hogares_final.dta")

# Kernel uniforme
def kernel_uniform(x, data, h):
    n = len(data)
    dens = np.zeros_like(x)
    for xi in data:
        mask = np.abs(x - xi) <= h
        dens[mask] += 1 / (2 * h)
    return dens / n

# Años a analizar
años = [2004, 2024]

for año in años:
    df_año = df[df["ano4"] == año].copy()

    # Recorte de outliers: entre percentil 1 y 99
    p1 = df_año["ingreso_hogar"].quantile(0.01)
    p99 = df_año["ingreso_hogar"].quantile(0.99)
    df_año = df_año[(df_año["ingreso_hogar"] >= p1) & (df_año["ingreso_hogar"] <= p99)]

    # Separar grupos
    ocup = df_año[df_año["algun_desocupado"] == 0]["ingreso_hogar"]
    desocup = df_año[df_año["algun_desocupado"] == 1]["ingreso_hogar"]

    # Bandwidth óptimo para kernel uniforme
    n = len(df_año)
    std = df_año["ingreso_hogar"].std()
    h_opt = 2.34 * std * n ** (-1/5)

    # Rango de evaluación
    x_vals = np.linspace(df_año["ingreso_hogar"].min(), df_año["ingreso_hogar"].max(), 500)

    # Densidades
    dens_ocup = kernel_uniform(x_vals, ocup.values, h_opt)
    dens_desocup = kernel_uniform(x_vals, desocup.values, h_opt)

    # Gráfico
    plt.figure(figsize=(10, 5))
    plt.plot(x_vals, dens_ocup, label="Todos ocupados", color="#9467bd", linewidth=2.2)  # violeta
    plt.plot(x_vals, dens_desocup, label="Al menos un desocupado", color="#ff7f0e", linewidth=2.2)  # naranja
    plt.title(f"Densidad de ingreso por hogar (kernel uniforme óptimo) - Año {año}")
    plt.xlabel("Ingreso del hogar")
    plt.ylabel("Densidad estimada")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Cargar base
df = pd.read_stata("hogares_final.dta")

# Funciones de kernels
def kernel_gaussiano(x, data, h):
    n = len(data)
    u = (x[:, None] - data[None, :]) / h
    dens = np.mean((1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * u**2), axis=1) / h
    return dens

def kernel_epanechnikov(x, data, h):
    n = len(data)
    u = (x[:, None] - data[None, :]) / h
    mask = np.abs(u) <= 1
    dens = np.mean(0.75 * (1 - u**2) * mask, axis=1) / h
    return dens

# Años a comparar
años = [2004, 2024]

for año in años:
    df_año = df[df["ano4"] == año].copy()

    # Recorte de outliers (percentil 1 y 99)
    p1 = df_año["ingreso_hogar"].quantile(0.01)
    p99 = df_año["ingreso_hogar"].quantile(0.99)
    df_año = df_año[(df_año["ingreso_hogar"] >= p1) & (df_año["ingreso_hogar"] <= p99)]

    # Separar grupos
    ocup = df_año[df_año["algun_desocupado"] == 0]["ingreso_hogar"].values
    desocup = df_año[df_año["algun_desocupado"] == 1]["ingreso_hogar"].values

    # Bandwidth óptimo (usamos mismo criterio que antes)
    n = len(df_año)
    std = df_año["ingreso_hogar"].std()
    h = 2.34 * std * n ** (-1/5)

    # Rango de evaluación
    x_vals = np.linspace(df_año["ingreso_hogar"].min(), df_año["ingreso_hogar"].max(), 500)

    # Estimar con kernel gaussiano
    dens_g_ocup = kernel_gaussiano(x_vals, ocup, h)
    dens_g_desocup = kernel_gaussiano(x_vals, desocup, h)

    # Estimar con kernel epanechnikov
    dens_e_ocup = kernel_epanechnikov(x_vals, ocup, h)
    dens_e_desocup = kernel_epanechnikov(x_vals, desocup, h)

    # Graficar Gaussiano
    plt.figure(figsize=(10, 5))
    plt.plot(x_vals, dens_g_ocup, label="Todos ocupados", color='#9467bd')       # violeta
    plt.plot(x_vals, dens_g_desocup, label="Al menos un desocupado", color='#ff7f0e')  # naranja
    plt.title(f"Densidad - Kernel Gaussiano - Año {año}")
    plt.xlabel("Ingreso del hogar")
    plt.ylabel("Densidad estimada")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    # Graficar Epanechnikov
    plt.figure(figsize=(10, 5))
    plt.plot(x_vals, dens_e_ocup, label="Todos ocupados", color='#9467bd')       # violeta
    plt.plot(x_vals, dens_e_desocup, label="Al menos un desocupado", color='#ff7f0e')  # naranja
    plt.title(f"Densidad - Kernel de Epanechnikov - Año {año}")
    plt.xlabel("Ingreso del hogar")
    plt.ylabel("Densidad estimada")
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()
```
